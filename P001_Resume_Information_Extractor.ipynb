```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
resume_text = """ 

Sneh Patel
AI Engineer | DL Engineer
Phone : +91-8733806962
Email : snehmpatel20@gmail.com

Profile : 
I am an AI & Machine Learning specialist with 2.5 years of deep experience in building intelligent systems. Currently, I’m solving complex data problems at [Current Company], but I’m looking for a bigger challenge where I can lead larger deployments.

Skills : 
Python, Java, NLP, ML, DL

Education :
B.tech in Information technology, GNU (2023) 

"""
```

```python
import re 

def extract_email(text):
    email_pattern = r'\S+@\S+'

    match = re.search(email_pattern, text)

    if match:
        return match.group()
    else: 
        return None
    

email = extract_email(resume_text)
print(f"Found Email: {email}")
```

```python
def extract_name(text):
    doc = nlp(text)

    for ent in doc.ents:
        if ent.label_ == "PERSON":
            return ent.text
    return "Name Not Found"

name = extract_name(resume_text)
print(f"Found Name: {name}")
```

```python
def extract_phone(text):

    phone_number = r'(\+?\d{1,3}[-.\s]?)?(\(?\d{3}\)?[-.\s]?)?\d{3}[-.\s]?\d{4}'

    match = re.search(phone_number, text)

    if match:
        return match.group()
    else:
        return None 
    
phone = extract_phone(resume_text)
print(f"Found Phone: {phone}")
```

```python
def parse_resume(text):
    data = {
        "Name" : extract_name(text),
        "Email" : extract_email(text),
        "Phone" : extract_phone(text)
    }

    return data
```

```python
final_output = parse_resume(resume_text)
```
