```python
import pandas as pd 

documents = [
    # Doc 1: About Python
    "Python is an amazing programming language. Python is great for Data Science.",
    
    # Doc 2: About Healthy Eating
    "Apples and bananas are healthy fruits. Eating fruit is good for your health.",
    
    # Doc 3: About Space
    "NASA plans to send a rocket to Mars. Space exploration is the future."
]

titles = ["Tech Blog", "Health Blog", "Space Blog"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

feature_names = tfidf_vectorizer.get_feature_names_out()

print("Shape:", tfidf_matrix.shape)
```

```python
def extract_top_keywords(doc_index):

    row = tfidf_matrix[doc_index]

    dense_row = row.todense().tolist()[0]

    keywords = []
    for i, score in enumerate(dense_row):
        if score > 0:
            word = feature_names[i]
            keywords.append((score,word))

    keywords.sort(key=lambda x:x[0], reverse=True)

    return keywords[:5]

top_words = extract_top_keywords(0)

print(f"--- Top Keywords for '{titles[0]}' ---")
for score, word in top_words:
    print(f"{word}: {score:.4f}")
```

```python
for i in range(len(documents)):
    print(f"\nðŸ“„ Document: {titles[i]}")

    top_words = extract_top_keywords(i)

    for score, word in top_words:
        print(f"   ðŸ”¹ {word} ({score:.2f})")
```
