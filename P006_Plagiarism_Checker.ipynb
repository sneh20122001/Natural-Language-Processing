```python
documents = [
    "The universe is vast and filled with billions of stars and galaxies.",   
    "The universe is huge and full of billions of stars and galaxies.",       
    "I love playing football and watching sports on the weekend."             
]

student_names = ["Student A", "Student B", "Student C"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()

tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

print("Matrix Shape:", tfidf_matrix.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(tfidf_matrix)

print("Raw Similarity Matrix:\n", similarity_matrix)
```

```python
reference_index = 0 

print(f"--- Plagiarism Report for {student_names[reference_index]} ---\n")

for i in range(len(documents)):
    if i == reference_index:
        continue 

    score = similarity_matrix[reference_index][i]
    percent = score * 100 

    print(f"Vs {student_names[i]}: {percent:.2f}% Similarity")

    if percent > 70:
        print("   ðŸš¨ PLAGIARISM DETECTED! \n")
    else:
        print("   âœ… Content is Unique. \n")
```

```python
def check_plagiarism(text1, text2):

    docs = [text1, text2]

    vec = TfidfVectorizer()
    matrix = vec.fit_transform(docs)
    
    similarity = cosine_similarity(matrix[0:1], matrix[1:2])
    
    return similarity[0][0]

text_original = "Machine learning is a field of AI."
text_copied = "AI contains a field called Machine Learning."

score = check_plagiarism(text_original, text_copied)
print(f"Similarity Score: {score * 100:.2f}%")
```
